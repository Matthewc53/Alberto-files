```python
weight_list = []

for index, row in df.iterrows():
    # Split the 'To' column by commas
    weights = row['D. Zip Ranges'].split(',')
    for (weight) in (weights):
        new_row = {
            'Dest': weight,
            'DISCOUNT %' : row['DISCOUNT %'],
            'MINIMUM CHARGE' : row['MINIMUM CHARGE'],            
            
            

        }
        
        weight_list.append(new_row)
```

```python
Class_list = []

for index, row in df.iterrows():
    # Split the 'To' column by commas
    classes = row['cl_1'].split(',')
    for (clas) in (classes):
        new_row = {
            'Class': clas,
            'TQL PO':row['TQL PO']
            
            
}
        
        Class_list.append(new_row)
```

```python
Dims_list = []

for index, row in df.iterrows():
    # Split the 'To' column by commas
    dims = row['dim_1'].split(',')
    for (dim) in (dims):
        new_row = {
            'TQL PO':row['TQL PO'],
            'Dimensions': dim,
            
        }
        
        Dims_list.append(new_row)
```

```python
class_df = pd.DataFrame(Class_list)
```

```python
df = df2_subs

# Function to find matching column header
def find_matching_column(row):
    lc_value = row['LC']
    
    for col in df.columns:
        # Skip LC column itself if needed
        if col == 'LC':
            continue
        
        if row[col] == lc_value:
            return col
    
    return None  # if no match found

# Apply function to each row
df['SCAC'] = df.apply(find_matching_column, axis=1)

# Save output
df.to_csv("output_with_match.csv", index=False)
```

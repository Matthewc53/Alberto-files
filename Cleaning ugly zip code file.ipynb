```python
import pandas as pd
import numpy as np
import os
```

```python

csv = pd.read_excel('Zip Code list (NTG LTL Quote)_Kelly Dawson.xlsx')

df = pd.DataFrame(csv)
```

```python
df
```

```python
df.rename(columns=df.iloc[0], inplace = True)
```

```python
df
```

```python
df.dropna(axis=1, how='all')
```

```python
df = df.apply(pd.to_numeric,errors='coerce')
```

```python
df = df.dropna(axis=0,how='all')
```

```python
df = df.dropna(axis=1,how='all')
```

```python
df
```

```python
pt_1 = df.iloc[:,0]

pt_2 = df.iloc[:,1]

pt_3 = df.iloc[:,2]
```

```python
new_df = pd.concat([pt_1, pt_2, pt_3])
```

```python
clean = pd.DataFrame(new_df)
```

```python
clean = clean.dropna(how='any')
```

```python
clean.to_csv('cleanzips.csv')
```

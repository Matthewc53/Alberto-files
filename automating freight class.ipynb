# Find closest class in Average class

```python
import pandas as pd
import numpy as np

def find_closest_value(target, value_list):
    """
    Finds the closest value in a list to a given target value.
    """
    if not value_list:
        return None  # Handle empty list case
    
    # Calculate absolute differences and find the index of the minimum difference
    closest_index = np.argmin(np.abs(np.array(value_list) - target))
    return value_list[closest_index]



# Define the list of values to compare against
comparison_list = classes

# Apply the function to the DataFrame column
df['Closest_Class'] = df['Average Class'].apply(lambda x: find_closest_value(x, comparison_list))
```

# Picking class subs for loop

```python
emplist = []

for x in df['Density']:
    if (x<1):
        emplist.append(400)
    elif (x>=1) & (x<2):
        emplist.append(300)
    elif (x>=2) & (x<4):
        emplist.append(250)
    elif (x>=4) & (x<6):
        emplist.append(175)
    elif (x>=6) & (x<8):
        emplist.append(125)   
    elif (x>=8) & (x<10):
        emplist.append(100)
    elif (x>=10) & (x<12):
        emplist.append(92.5)
    elif (x>=12) & (x<15):
        emplist.append(85)
    elif (x>=15) & (x<22.5):
        emplist.append(70)
    elif (x>=22.5) & (x<30):
        emplist.append(65) 
    elif (x>30):
        emplist.append(60) 
    else: 
        emplist.append("Not Classed")
```

```python
import pandas as pd
import numpy as np

input_file = df   # change to your file name
output_file_wgt = "output_weights.xlsx"


# function to collect weights from each row
def collect_weights(row):
    weights = []
    for col in df.columns:
        if col.startswith("Average of Weight"):
            val = row[col]
            if pd.notna(val):
                weights.append(val)
    return weights

# create list of weights per row
df["all_weights"] = df.apply(collect_weights, axis=1)

# group by shipmentID and combine lists
grouped_wgt = df.groupby("Lane")["all_weights"].sum().reset_index()

# find max number of weights in any shipment
max_len = grouped_wgt["all_weights"].apply(len).max()

# expand into separate columns
for i in range(max_len):
    grouped_wgt[f"weight_{i+1}"] = grouped_wgt["all_weights"].apply(
        lambda x: x[i] if i < len(x) else np.nan
    )

# drop helper column
grouped_wgt = grouped_wgt.drop(columns=["all_weights"])

# save to Excel
grouped_wgt.to_excel(output_file_wgt, index=False)

```

```python
import pandas as pd
import numpy as np

input_file = df   # change to your file name
output_file_cl = "output class.xlsx"


# function to collect weights from each row
def collect_weights(row):
    weights = []
    for col in df.columns:
        if col.startswith("Class"):
            val = row[col]
            if pd.notna(val):
                weights.append(val)
    return weights

# create list of weights per row
df["all_Class"] = df.apply(collect_weights, axis=1)

# group by shipmentID and combine lists
grouped_cl = df.groupby("Lane")["all_Class"].sum().reset_index()

# find max number of weights in any shipment
max_len = grouped_cl["all_Class"].apply(len).max()

# expand into separate columns
for i in range(max_len):
    grouped_cl[f"class_{i+1}"] = grouped_cl["all_Class"].apply(
        lambda x: x[i] if i < len(x) else np.nan
    )

# drop helper column
grouped_cl = grouped_cl.drop(columns=["all_Class"])

# save to Excel
grouped_cl.to_excel(output_file_cl, index=False)

```

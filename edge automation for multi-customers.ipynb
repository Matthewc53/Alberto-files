```python
df = pd.read_excel(latest_file, sheet_name='Sheet1')

#df1 = pd.read_excel(latest_file, sheet_name='Product Details')
```

```python
csts = df['Customer Code'].value_counts().index.tolist()
```

```python
csts
```

```python
cs_one = csts[0]

cs_two = csts[1]

cs_three = csts[2]

cs_four = csts[3]

cs_five = csts[4]
```

```python
cs_one
```

```python
df1 = df[df['Customer Code']==cs_one]
df2= df[df['Customer Code']==cs_two]
df3= df[df['Customer Code']==cs_three]
df4= df[df['Customer Code']==cs_four]
df5= df[df['Customer Code']==cs_five]
```

```python
df1_col = df1.columns.tolist()
```

```python
df_filt = df1.columns.to_frame()
```

```python
df_filt = df_filt[df_filt[0].str.contains('Acc|Pro|Pr|Dest Zip|Weight|Class|Origin Zip',na=False)]
```

```python
df_filt = df_filt.sort_values(by=[0])
```

```python
df_filt = df_filt[0].to_list()
```

```python
df_filt
```

```python
df1 = df1.filter(df1.filter(items=df_filt))

df2 = df2.filter(df2.filter(items=df_filt))

df3 = df3.filter(df3.filter(items=df_filt))

df4 = df4.filter(df4.filter(items=df_filt))

df5 = df5.filter(df5.filter(items=df_filt))
```
